## Предобработка данных

```python
#приведем название колонки к правильному регистру
df = df.rename(columns={'cityCenters_nearest': 'city_centers_nearest'})
# Приводим к типу даты дату публикации
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
# Выясняем количество значений в аппартаментах. Всего аппартаменты - 50. Заменяем пропущенные значения на нет
df['is_apartment'] = df['is_apartment'].fillna(value=False)
# Приводим в целочисленный тип количество балконов. Заменяем пропущенные значения на 0
df['balcony'] = df['balcony'].fillna(value=0).astype(int)
# Приводим в целочисленный тип количество парков. Заменяем пропущенные значения на 0
df['parks_around3000'] = df['parks_around3000'].fillna(value=0).astype(int)
# Приводим в целочисленный тип количество прудов. Заменяем пропущенные значения на 0
df['ponds_around3000'] = df['ponds_around3000'].fillna(value=0).astype(int)

# Смотрим количество пропущенных значений
df.isnull().sum()
```

**city_centers_nearest не будем трогать, тк считать медианой будет ошибкой, ведь расстояние до центра сильно зависит от размера населенного пункта, но по имеющимся данным восстановить эти данные нельзя. Поэтому эти пропуски трогать пока не будем
Это также касается airports_nearest по той же логике. Ситуаций, когда нет одного, но есть другое очень мало.
Также не будем трогать days_exposition предполагая, что если нет данных, объявление не закрыто. Для дальнейшего анализа эти данные нам не понадобятся**

```python
# Удаляем записи, где отсутствует локация и этажность, тк записей с пустым locality_name немного, а сгенерировать мы их никак не можем, тоже самое с floors_total
df.dropna(subset=['locality_name'], inplace=True)
df.dropna(subset=['floors_total'], inplace=True)

# Приводим этажность к целочисленному типу
df['floors_total'] = df['floors_total'].astype(int)

# Заменяем оставшиеся пустые значения высоты потолков на медианное значение по всему датасету
df['ceiling_height'] = df.groupby(['locality_name', 'floors_total'])['ceiling_height'].apply(lambda x: x.fillna(x.median()))
df['ceiling_height'].fillna(df['ceiling_height'].median(), inplace=True)
```

```python
# Заменяем пустые значения жилой площади и площади кухни
living_ratio = df['living_area'].mean() / df['total_area'].mean()
kitchen_ratio = df['kitchen_area'].mean() / df['total_area'].mean()
df['living_area'].fillna(living_ratio * df['total_area'], inplace=True)
df['kitchen_area'].fillna(kitchen_ratio * df['total_area'], inplace=True)
```

```python
# Смотрим количество пропущенных значений
df.isnull().sum()
```

**Мы уже заменили пропущенные значения парков и водоемов на 0, также чаще всего не указано расстояние для случая, когда парка в трех километрах нет, следовательно расстояние до ближайшего парка превышает это значение. Заполним в таком случае расстояние 3500 метров**

```python
df['parks_nearest'] = df['parks_nearest'].fillna(3500)
df['ponds_nearest'] = df['ponds_nearest'].fillna(3500)
```

```python
# Смотрим количество пропущенных значений
df.isnull().sum()
```

Выяснили, что некоторые значения не имеется возмодности заполнить, поэтому оставили их как есть, хотя следует сообщить об этом тому, кто выгружал данные. ## Расчёты и добавление результатов в таблицу

```python
#Добавляем столбец со стоимостью квадратного метра
df['square_meter_price'] = round(df['last_price'] / df['total_area'], 2)

# Добавим день недели, месяц и год публикации объявления;
df['day_exposition'] = df['first_day_exposition'].dt.day
df['month_exposition'] = df['first_day_exposition'].dt.month
df['year_exposition'] = df['first_day_exposition'].dt.year
# Создаём функцию для заполнения нового столбца с типом этажа(первый, последний, другой)
def get_floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

df['floor_type'] =  df.apply(get_floor_type, axis=1)

# Добавим соотношение жилой и общей площади, а также отношение площади кухни к общей.
df['area_ratio'] = round(df['living_area'] / df['total_area'], 2)
df['kitchen_ratio'] = round(df['kitchen_area'] / df['total_area'], 2)

# проверим новые поля
df.head(10)
```

-	этаж квартиры; варианты — первый, последний, другой;

3)Необходимо убрать редкие и выбивающиеся значения.

```python
# Функция для подсчёта границ с учетом 1.5 размаха
def quartile_range(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq

# Столбцы с выбросами
list_emissions = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'days_exposition', 'ceiling_height', 'rooms']
for column in list_emissions:
    df[column] = quartile_range(df, column)

# Возвращаем в целочисленный тип число комнат
df['rooms'] = df['rooms'].fillna(0).astype(int)

# Изучаем, есть ли выбивающиеся значения
df.describe()
```
